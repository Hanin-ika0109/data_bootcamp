### Creating Dataframe

```python
import pandas as pd
import numpy as np
```

```python
# Creating Dataframe
# Python List [ ... ]
list1 = ["Swimming", "Cycling", "Hiking"]
list2 = [10000, 100, 400000]

dataframe_list = pd.DataFrame(list(zip(list1,list2)), #zip = pengelompokkan data
                              columns=['Hobby','Price'])
dataframe_list
```

```python
# Creating Dataframe
# Python Tupple ( ... )
tupple1 = ("Machine Learning", 2)
tupple2 = ("Marketing", 2)

dataframe_tupple = pd.DataFrame([tupple1,tupple2], 
                                columns=['Object','Total'])
dataframe_tupple
```

```python
# Creating Dataframe
# Numpy Array
array = np.array([["Meatball",5,7.7], ["Fried Rice",3,8.0],["Pizza", 10.0,9.0]])

dataframe_array = pd.DataFrame(array, 
                              columns=["Food","Price (USD)", "Rate"])
dataframe_array
```

```python
# Creating Dataframe
# Pandas Series ([ ... ])
series1 = pd.Series(["House A", "House B", "House C"])
series2 = pd.Series([100000, 200000, 300000])

series_dataframe = pd.DataFrame({"Type of House":series1, "Price": series2})
series_dataframe
```

### Sorting Dataframe

```python
# reading csv file 
data = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/Human%20Capital.csv")
print(data.shape)
data.head(5)
```

```python
# random sample data, n=10
data_10 = data.sample(n=10)
data_10
```

```python
# basic sorting - Alphabetic
data_sort = data.sort_values(by=['department'])
data_sort
```

```python
# basic sorting - Alphabetic Descending
data.sort_values(by=['department'], ascending=False)
```

```python
# basic sorting - Numerical
data.sort_values(by=['age'])
```

```python
# sorting more than two column
data.sort_values(by=['department','age'])
```

```python
# sorting data and show one column
data['employee_id'].sort_values(ascending=False)
```

### Filtering in Dataframe

```python
data_filter = data_10 = data.sample(n=10)
print(data_filter.shape)
data_filter
```

```python
# filtering beberapa kolom
data[["department","awards_won"]]
```

```python
# filtering menggunakan loc (untuk melakukan seleksi kolom dan baris yang berdasar pada nama baris dan kolom)
data.loc[(data["department"]== "Sales & Marketing")].head(3)
```

```python
# filter menggunakan iloc (untuk melakukan seleksi kolom dan baris yang berdasar indeks baris dan kolom)
data.iloc[0:5,0:3] # [row(index), column(index)]
```

```python
# filter data dengan memunculkan usia > 20 tahun dan length of service > 3 
data.loc[(data["age"] > 20) & (data["length_of_service"] > 3)].head(10)
```

### Creating New Variable

```python
data.head()
```

```python
data['others'] = data['age'] + data['length_of_service']
data.head()
```

```python
data['penggabungan'] = str(data['department']) + str(data['region'])
data.head()
```

### Grouping the Data

```python
# grouping average age and median in region variable
data.groupby('region').agg(jumlah_usia = ('age','mean'),
                           max_service = ('length_of_service','max')).head(5)
```

### Merging the Data

```python
data1 = data.iloc[:20,:6]
data1.head()
```

```python
data2 = data.loc[18:30, ['employee_id','is_promoted','no_of_trainings']]
data2.head()
```

```python
# merging data - left join 
data1.merge(data2, how='left', on='employee_id')
```

### Indexing Dataframe

```python
data = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/insurance.csv")
data
```

```python
# reset index starting from 0 
data.reset_index(drop=True)
```

```python
# set column as index > 2 column
data.set_index(['sex','smoker']).head()
```

```python
# set column as index 1 column
data.set_index('sex').head()
```

### Dropping Column

```python
# dropping column
# dropping column one column 
x = data.drop('charges', axis=1)
y = data[['charges']]

# dropping more than 1 column
data.drop(['age','sex'], axis=1)
```

### Join Dataframe

```python
# join dataframe
data_1 = data.sample(n=5)
data_2 = data.sample(n=9)

display(data_1.head())
display(data_2.head())
```

```python
# join dataframe (cont.)
data_1.join(data_2, lsuffix='_first', rsuffix='_second')
```

### Concatenate & Append Dataframe

```python
# concatenate - Horizontal
pd.concat([data_1, data_2], axis=1)
```

```python
# concatenate - Vertical
pd.concat([data_1, data_2], axis=0)
```

```python
# append 
data_1.append(data_2)
```

### Pivot Table Dataframe

```python
# pivot table
pd.pivot_table(data, values='age', index=['sex','smoker'], columns='region', aggfunc=['mean','min'])
```

### Challenge Time! Part 2

```python
data = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/insurance.csv")
data.head()
```

```python
data['bmi_categ_lambda'] = data['bmi'].apply(lambda x: 'High BMI' if x >= 50 else 'Medium BMI' if x >= 30 and x < 50 else 'Low BMI')
data.head()
```

```python
pd.pivot_table(data, values='bmi', index=['bmi_categ_lambda'], aggfunc=['min','max','mean'])
```

```python
data_upper = data[['sex','region','bmi_categ_lambda']]=data[['sex','region','bmi_categ_lambda']].apply(lambda col: col.str.upper())
data_upper.head()
```
